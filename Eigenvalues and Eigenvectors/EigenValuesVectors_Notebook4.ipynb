<b>Question:</b> <br>
Let us consider an internet with 4 pages, namely page1, page2, page3, and page4, which refer to each other. The problem is that we have to find the rank of each page using linear algebra.

<b> Solution: </b><br>
In order to solve the problem we have follow these steps,<br>
1. Construct a model
2. Represent the model with suitable data structure
3. Build a linear algebraic equation
4. Solve the equation algorithmically

## 1. Construct a Directed Graph Model
Outgoing edges indicate the pages referred to by a particular page, while incoming edges originate from pages referring to that particular page.

image.png

## 2. Build a probabilistic matrix
Each element in the matrix represents the probability of transitioning from the page represented by the ith row to the page represented by the jth column.

```python
import numpy as np
```

```python
A = np.array([[0, 0, 1, 1/2],
              [1/3, 0, 0, 0],
              [1/3, 1/2, 0, 1/2],
              [1/3, 1/2, 0, 0]])
```

image.png

```python
def eigen_values_vectors(M):
    # Calculate eigenvalues and eigenvectors
    eigenvalues, eigenvectors = np.linalg.eig(M)

    # Print eigenvalues and eigenvectors
    print("Eigenvalues:\n", eigenvalues)
    print("\nEigenvectors:")

    # Concatenate eigenvectors into a matrix
    eig_matrix = np.column_stack(eigenvectors)
    print(eig_matrix)

    # Extract the real part of eigenvalues
    real_parts = np.real(eigenvalues)

    # Find the index where the real part is closest to 1
    index_of_1 = np.argmin(np.abs(real_parts - 1))
    return np.real(eig_matrix[index_of_1])
```

```python
def scale_vect(V):
    return V/sum(V)
```

```python
pagerank = eigen_values_vectors(A)
```

```python
print(pagerank)
```

###Scaling the Eigen Vectors

```python
scaled_pagerank = scale_vect(pagerank)
pages = ['a', 'b', 'c', 'd']
print("Importance of each page:")
for i, j in zip(pages, scaled_pagerank):
    print(i," = ", j)
```
